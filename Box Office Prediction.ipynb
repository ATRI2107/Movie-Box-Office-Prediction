```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv',index_col=3)
credits=pd.read_csv('tmdb_5000_credits.csv',index_col=0)
```

```python
movies.head(2)
```

```python
movies['profit']=movies.revenue-movies.budget
movies.head(2)
```

```python
credits.head(2)
```

<h1>2. Data Cleaning </h1>

```python
credits.drop(['title'],axis=1,inplace=True)
```

```python
credits.head(2)
```

```python
df=movies.join(credits)
df.head(2)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df['lang_encoded']=labelencoder.fit_transform(df['original_language'])
```

```python
df=df[df['budget']>0]
```

```python
df.dropna(inplace=True)
df.head()
```

```python
from pandas.io.json import json_normalize
import json
def json_decode(data,key):
    result=[]
    data=json.loads(data)
    for item in data:
        result.append(item[key])
    return result
```

```python
df.genres=df.genres.apply(json_decode,key='name')
df.keywords=df.keywords.apply(json_decode,key='name')
df.production_companies=df.production_companies.apply(json_decode,key='name')
df.production_countries=df.production_countries.apply(json_decode,key='name')
df.cast=df.cast.apply(json_decode,key='name')
df.crew=df.crew.apply(json_decode,key='name')
df.spoken_languages=df.spoken_languages.apply(json_decode,key='name')
```

```python
df.head(2)
```

<h1>3. Data Visualization </h1>

<h3>3.1 Word Cloud</h3>

```python
from wordcloud import WordCloud
genres={}
for i in df['genres']:
    for j in i:
        if j not in genres:
            genres[j]=1
        else:
            genres[j]+=1
plt.figure(figsize=(12,8))
wordcloud=WordCloud(background_color="white",width=1000,height=1000,max_words=10,relative_scaling=0.5,normalize_plurals=False).generate_from_frequencies(genres)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

<h3>3.2 Letter Value Plots</h3>

```python
plt.figure(figsize=(15,11))
g1=sns.boxenplot(x='original_language',y='revenue',data=df[(df['original_language'].isin((df['original_language'].sort_values().value_counts()[:10].index.values)))])
g1.set_title("Revenue by Language",fontsize=20)
g1.set_xticklabels(g1.get_xticklabels(),rotation=45)
g1.set_xlabel("Language",fontsize=18)
g1.set_ylabel("Revenue",fontsize=18)
plt.show()
```

<h3>3.3 Heatmap</h3>

```python
col=['revenue','budget','popularity','runtime','lang_encoded']
plt.subplots(figsize=(10,8))
corr=df[col].corr()
sns.heatmap(corr,xticklabels=col,yticklabels=col,linewidths=.5,cmap="Reds")
```

<h1>4. Machine Learning Mechanisms and comparison of the Results<h1>

```python
x=df[['budget','popularity','runtime','lang_encoded']]
y=df[['revenue']]
```

```python
from sklearn.preprocessing import MinMaxScaler
x=MinMaxScaler().fit_transform(x)
y=MinMaxScaler().fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=13,test_size=0.2)
```

<h2> Training and testing of data set on the following algorithms </h2>

<h3> 1. Linear Regression </h3>

```python
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
model=regr.fit(x_train,y_train)
model.score(x_test,y_test)
```

<h3> 2. Random Forest</h3>

```python
from sklearn.ensemble import RandomForestRegressor
regr=RandomForestRegressor(random_state=13,n_estimators=500)
model=regr.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.svm import SVR
regr=SVR()
model=regr.fit(x_train,y_train)
model.score(x_test,y_test)
```
